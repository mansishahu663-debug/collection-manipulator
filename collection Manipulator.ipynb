```python
students = []

while True:
    print("\n1. Add Student")
    print("2. Display All Students")
    print("3. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        student_id = input("Student ID: ")
        name = input("Name: ")
        age = input("Age: ")
        grade = input("Grade: ")
        dob = input("Date of Birth (YYYY-MM-DD): ")
        subjects = input("Subjects (comma-separated): ")

        student = {
            "ID": student_id,
            "Name": name,
            "Age": age,
            "Grade": grade,
            "DOB": dob,
            "Subjects": subjects
        }

        students.append(student)
        print("Student added successfully!")

    elif choice == "2":
        print("\nStudent List")
        for s in students:
            print(
                f"ID: {s['ID']} | Name: {s['Name']} | "
                f"Age: {s['Age']} | Grade: {s['Grade']} | "
                f"Subjects: {s['Subjects']}"
            )

    elif choice == "3":
        print("Goodbye!")
        break

    else:
        print("Invalid choice!")
```
